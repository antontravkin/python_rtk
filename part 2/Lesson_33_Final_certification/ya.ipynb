```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import shap
import zipfile
import missingno as msno
import lightgbm as lgb
from lightgbm import LGBMClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from category_encoders import TargetEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import f1_score, accuracy_score, roc_auc_score

# === Seed
np.random.seed(42)

# === Load data
with zipfile.ZipFile('playground-series-s5e7.zip', 'r') as zip_ref:
    zip_ref.extractall('./')

df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')

# === Feature Engineering


def create_features(df):
    df['alone_ratio'] = df['Time_spent_Alone'] / (
        df['Time_spent_Alone'] + df['Social_event_attendance'] +
        df['Going_outside'] + 1
    )
    df['social_ratio'] = df['Social_event_attendance'] / \
        (df['Going_outside'] + 1)
    df['is_high_poster'] = (df['Post_frequency'] > 5).astype(int)
    return df


df_train = create_features(df_train)
df_test = create_features(df_test)

# === Target
target = 'Personality'
label_encoder = LabelEncoder()
df_train['target'] = label_encoder.fit_transform(df_train[target])

# === Features
num_features = [
    'Time_spent_Alone', 'Social_event_attendance', 'Going_outside',
    'Friends_circle_size', 'Post_frequency', 'alone_ratio', 'social_ratio'
]
cat_features = ['Stage_fear', 'Drained_after_socializing', 'is_high_poster']
features = num_features + cat_features

# === Preprocessor
preprocessor = ColumnTransformer([
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler())
    ]), num_features),
    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('target_encoder', TargetEncoder())
    ]), cat_features)
])

# === Model Pipeline
model = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', LGBMClassifier(random_state=42))
])

# === Cross-Validation and Threshold Optimization
X = df_train[features]
y = df_train['target']
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
best_thresholds = []
val_scores = []

for fold, (train_idx, val_idx) in enumerate(skf.split(X, y)):
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    # Fit model
    model.fit(X_train, y_train)

    # Predict probabilities
    # Assuming binary classification
    probas_val = model.predict_proba(X_val)[:, 1]

    # Optimize threshold
    thresholds = np.arange(0.1, 0.9, 0.01)
    f1_scores = [f1_score(y_val, (probas_val >= thr).astype(int))
                 for thr in thresholds]
    best_thr = thresholds[np.argmax(f1_scores)]
    best_thresholds.append(best_thr)

    # Validation score
    preds_val = (probas_val >= best_thr).astype(int)
    val_scores.append({
        'f1': f1_score(y_val, preds_val),
        'accuracy': accuracy_score(y_val, preds_val),
        'roc_auc': roc_auc_score(y_val, probas_val)
    })
    print(f"Fold {fold+1}: F1={val_scores[-1]['f1']:.4f}, "
          f"Accuracy={val_scores[-1]['accuracy']:.4f}, "
          f"ROC AUC={val_scores[-1]['roc_auc']:.4f}, "
          f"Best Threshold={best_thr:.3f}")

# === Final Model and Submission
final_thr = np.mean(best_thresholds)
print(f"\nFinal averaged threshold: {final_thr:.3f}")

# Fit model on full training data
model.fit(X, y)

# Predict on test set
X_test = df_test[features]
probas = model.predict_proba(X_test)[:, 1]  # Assuming binary classification
preds_final = (probas >= final_thr).astype(int)
df_test['Personality'] = label_encoder.inverse_transform(preds_final)
df_test[['id', 'Personality']].to_csv('submission.csv', index=False)

print("âœ… Submission file created: submission.csv")

# === SHAP Analysis
preprocessor.fit(X, y)  # Fit preprocessor on training data
X_test_prep = preprocessor.transform(X_test)  # Transform test data
explainer = shap.Explainer(model.named_steps['classifier'], X_test_prep)
shap_values = explainer(X_test_prep)
shap.summary_plot(shap_values, feature_names=features, plot_type="bar")
shap.summary_plot(shap_values, feature_names=features)

# === EDA
print("\n--- EDA ---")
print(df_train.info())
print(df_train.describe())

plt.figure(figsize=(10, 4))
msno.matrix(df_train)
plt.title('Missing Values Matrix')
plt.show()

for col in cat_features + [target]:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, data=df_train)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()

df_train[num_features].hist(figsize=(14, 10), bins=30)
plt.suptitle('Histograms of Numerical Features')
plt.show()

plt.figure(figsize=(10, 8))
sns.heatmap(df_train[num_features].corr(),
            annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
