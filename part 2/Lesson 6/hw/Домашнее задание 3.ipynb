#### *Решение*

```python
import sqlite3
import pandas as pd

# Загружаем CSV-файл
df = pd.read_csv('big_sales.csv', sep=';', encoding='utf-8')

# Чистим от пробелов и мусора (если нужно)
df.columns = df.columns.str.strip()
df['additional'] = df['additional'].astype(str).str.strip()

# Преобразуем NaN в None (SQLite-friendly)
df = df.where(pd.notnull(df), None)

# Подключаемся или создаём БД
conn = sqlite3.connect('big_sales.db')
cursor = conn.cursor()

# Создаём таблицу (если нет)
cursor.execute('''
    CREATE TABLE IF NOT EXISTS sales_data (
        date TEXT,
        sales REAL,
        category TEXT,
        additional TEXT
    )
''')

# Записываем данные в таблицу
df.to_sql('sales_data', conn, if_exists='append', index=False)

# Закрываем соединение
conn.commit()
conn.close()

print("Данные успешно импортированы в базу big_sales.db")
```
