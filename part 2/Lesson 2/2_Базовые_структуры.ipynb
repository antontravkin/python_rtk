### Список (List)

Список - это коллекция упорядоченных по индексам элементов. Является аналогом динамического массива из других языков программирования.

```python
my_list = list() # --> пустой список
my_list = [] # --> пустой список
my_list = [1,'string',3.1,1,1,True] # --> список [1,'string',3.1,1,1,True]
print(my_list)
print(my_list[0]) # --> 1 Обращение по индексу элемента
print(my_list[:5:2]) # --> [1, 3.1, 1] Слайсинг элементов
```

Методы списка

```python
my_list = [1, 2, 3, 4, 5, 6, 7, 8]
list2 = ['a', 'b', 'c', 'some_element']
index = 3
element = 'some_element'

print('Исходный список my_list: ', my_list)
print('Исходный список list2: ', list2)
my_list.sort(reverse=True) # Сортирует значения в списке
print('Сортировка my_list по убыванию: ', my_list)
my_list.append(element) # Добавляет элемент в конец списка
print(f'Список my_list после добавления в конец нового элемента: ({element})', my_list)
my_list.insert(index, element) # Вставляет в список элемент на указанный индекс
print(f'Список my_list после вставки элемента ({element}) на место указанного индекса ({index}): ', my_list)
my_list.remove(element) # Удаляет из списка первый найденный элемент
print('Список my_list после удаления указанного элемента (удаляется только первый по счету элемент): ', my_list)
el = my_list.pop(index) # Удаляет из списка элемент с указанным индексом и передает удаленное значение на экран.
print(el)
print('Список my_list после удаления указанного элемента по индексу: ', my_list)
print(f'Индекс указанного элемента ({element}):', my_list.index(element)) # Возвращает номер индекса первого найденного элемента в списке
my_list.extend(list2) # Добавляет список в конец исходного списка
print('Список my_list после добавления в него другого списка list2: ', my_list)
print(f'Количество вхождений указанного элемента ({element}): ', my_list.count(element)) # Кол-во вхождений определенного символа

# С помощью конструкции in можно проверить есть ли в списке искомый элемент
print(f'Есть ли элемент ({element}) в списке my_list: ', element in my_list )

my_list.clear()
print('Список my_list после очистки: ', my_list)
```

### Множество (Set)

Множество - это неупорядоченный набор уникальных объектов.
Удобен, когда нужно найти множесто уникальных элементов из коллекции.

image019-4.png

```python
# варианты создания
my_set = set()  # --> пустое множество
my_set2 = {1, 2, 3}  # --> множество из элементов {1, 2, 3}
my_set3 = {1, 2, 2, 3}  # --> множество из элементов {1, 2, 3}
print(my_set)
print(my_set2)
print(my_set3)
my_set3.add(4)  # --> добавляет 4 к множеству
print(my_set3)
el = my_set3.pop()  # --> достает из множества случайный элемент
print(el)
print(my_set3)
```

Операции над множеством

```python
my_set = {0, 1, 2, 3, 4, 5}
print('Мощность (количество элементов) множества: ', len(my_set))
my_set.add('element')
print('Добавить элемент в множество: ', my_set)
my_set.discard('element')
print('Удалить элемент из множества: ', my_set)
my_set.pop()
print('Удалить случайный элемент из множества: ', my_set)
```

Математические операции над множествами

```python
A = {1, 3, 5, 7}
B = {3, 7, 6, 4, 9, 10, 1}

print('Пересечение множеств: ', A & B)
print('Объединение множеств: ', A | B)
print('Симметричная разница: ', A ^ B)
print('Разница множеств A-B:', A - B)
print('Разница множеств B-A:', B - A)
```

### Словарь (Dict)

Хранилище данных типа ключ-значение. Близко к классической структуре данных HashTable.
* Например, может выступать как мини база данных. Размер и уровень вложенности не ограничен, хранить можно объекты любого типа.
* Ключи уникальны в пределах одного словаря.
* В качестве ключей могут выступать любые неизменяемые (immutable) встроенные типы.

```python
# варианты создания
my_dict = dict()  # --> пустой словарь {}
my_dict2 = {}  # --> пустой словарь {}
print(my_dict2)
ages = {
   "Dmitry": 26,
   "Alex": 34
}  # --> словарь с ключами "Dmitry" и "Alex"
print(ages)
del ages["Dmitry"]  # --> удалит ключ Dmitry из словаря
print(ages)
```

```python
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
print('Обращение по ключу "a": ', my_dict['a'])
my_dict['a'] = 't'
print('Поменять значение в словаре по ключу "a": ', my_dict)
my_dict['new_key'] = 'new_value'
print('Добавление новой пары ключ-значение: ', my_dict)
print('Проверить если ключ "b" есть в словаре: ', 'b' in my_dict)
print('Все ключи словаря, преобразованные в список: ', my_dict.keys())
print('Все значения словаря, преобразованные в список: ', my_dict.values())
print('Все значения словаря, преобразованные в список: ', my_dict.items())
```

### Генераторы

**Генераторы списков и словарей** - это сокращенные формы записи генерации соответствующих объектов. Они просты и лаконичны в использовании.

```python
lst = []
for i in range (10):
    lst.append(i)

print(lst)
```

Генератор списков

```python
# [выражение for итератор in коллекция]
lst1 = [i for i in range(10)]
lst2 = [s for s in 'hello world']
lst3 = [x**2 - 3*x + 7 for x in range(15)]
lst4 = [[0]*10 for i in range(5)] # таблица - двойной список из нулей, размером 5*10
lst5 = [(i, j) for i in 'abc' for j in [1, 2, 3]] # список кортежей вида [(a, 1), (a, 2), ..., (c, 3)]

print('lst1 =', lst1)
print('lst2 =', lst2)
print('lst3 =', lst3)
print('lst4 =', lst4)
print('lst5 =', lst5)
```

```python
lst = []
for i in range (10):
    if i%2 == 0:
        lst.append(i)

print(lst)
```

Генерация списков с условием

```python
# [выражение for итератор in коллекция if условие] - это еще называется list comprehension
lst5 = [i for i in range(10) if i%2 == 0]
lst6 = [i*j for i in [2, 3, 4, 5] for j in [1, 2, 3] if i*j > 10] # Так писать можно, но согласитесь, выглядит тяжело для прочтения

print('lst5 =', lst5)
print('lst6 =', lst6)
```

Генератор словарей

```python
dict0 = {}
for i in range(1,10):
    if i%2 == 0:
        dict0[i] = i**2

print(dict0)
```

Генератор словарей (с условием)

```python
dict1 = {i: i**2 for i in range(1, 10) if i%2 == 0}
dict2 = {s: ord(s) for s in 'hello world'} # ord возвращает индекс символа из таблицы символов ASCII

print('dict1 =', dict1)
print('dict2 =', dict2)
```

Рассмотрим простой пример пузырьковой сортировки чисел в списке - от меньшего к большему.

1. Создаем список из пяти чисел, расположенных случайным образом.
2. Внешний цикл запускается ровно столько раз сколько элементов в списке,инициируем запуск внутреннего цикла - цикла сравнения элементов.
3. В цикле сравнения для каждого элемента массива сравниваем этот элемент со стоящим справа от него.
Если элемент слева больше элемента справа, меняем их местами и переходим к следующей итерации внутреннего цикла.
4. После окончания всех итераций обоих циклов завершаем алгоритм.

Сколько оперций выполняет алгоритм пузырьковой сортировки?



* Big O называется так из-за того, что в математике "O" используется для обозначения "order of" (порядка) и позволяет сравнивать функции роста. Он представляет собой математическую нотацию, которая описывает, как алгоритм будет выполняться в наихудшем случае, исходя из размера входных данных.

* Big Theta (Θ): Big Theta также оценивает верхнюю и нижнюю границы временной сложности алгоритма, но описывает точную сложность, а не только наихудший случай, как Big O. Θ(f(n)) обозначает, что время выполнения алгоритма ограничено функцией f(n) как сверху, так и снизу.

* Big Omega (Ω): Big Omega оценивает нижнюю границу временной сложности алгоритма. Ω(f(n)) говорит о том, что алгоритм выполнится не быстрее, чем функция f(n).

* Little O (o): Little O представляет собой верхнюю границу, которая строже, чем Big O. Если f(n) является o(g(n)), это означает, что время выполнения алгоритма ограничивается функцией g(n), но алгоритм работает быстрее, чем g(n).

* Little Omega (ω): Little Omega представляет собой нижнюю границу, которая строже, чем Big Omega. Если f(n) является ω(g(n)), это означает, что алгоритм работает медленнее, чем g(n), но не медленнее, чем f(n).

```python
import time
import random
import string

# Создание тестовых данных
def generate_list(n):
    numbers = []
    for _ in range(n):
        number = random.randrange(1,100)
        numbers.append(number)
    return numbers

def generate_tuple(n):
    numbers = []
    for _ in range(n):
        number = random.randrange(1,100)
        numbers.append(number)
    numbers = tuple(numbers)
    return numbers

# Функции для поиска вхождений
def find_number(numbers, num):
    cnt = 0
    for number in numbers:
        if number == num: 
            cnt+=1 
    return cnt
```

```python
# Генерация данных
n = 10000000
number_list = generate_list(n)
number_tupl = generate_tuple(n)
```

```python
import time
import random
import string

# Создание тестовых данных
def generate_contacts_list(n):
    contacts = []
    for _ in range(n):
        name = ''.join(random.choices(string.ascii_letters, k=10))
        phone = ''.join(random.choices(string.digits, k=10))
        email = f"{name.lower()}@example.com"
        contacts.append([name, phone, email])
    return contacts

def generate_contacts_dict(n):
    contacts = {}
    for _ in range(n):
        name = ''.join(random.choices(string.ascii_letters, k=10))
        phone = ''.join(random.choices(string.digits, k=10))
        email = f"{name.lower()}@example.com"
        contacts[name] = {"phone": phone, "email": email}
    return contacts

# Функции для поиска контактов
def find_contact_list(contacts, name):
    for contact in contacts:
        if contact[0] == name:
            return contact
    return None

def find_contact_dict(contacts, name):
    return contacts.get(name, None)
```

```python
# Генерация данных
n = 100000
contacts_list = generate_contacts_list(n)
contacts_dict = generate_contacts_dict(n)
```

```python
contacts_list[:5]
```

```python
contacts_dict
```

```python
# Выбор случайного имени, которое мы будем искать с помощью дальнейших функции
random_name = contacts_list[random.randint(0, n-1)][0]
random_name
```

```python
%%time
# Время поиска в списке

start_time = time.time()
find_contact_list(contacts_list, random_name)
list_search_time = time.time() - start_time
list_search_time
```

```python
%%time
# Время поиска в словаре

start_time = time.time()
find_contact_dict(contacts_dict, random_name)
dict_search_time = time.time() - start_time
dict_search_time
```

#### Задача 1: Управление библиотекой книг

Создайте систему управления библиотекой, которая позволяет добавлять книги, удалять книги, искать книги по названию и автору.

```python
# Решение
# Список для хранения книг
books = []


def add_book(title: str, author: str):
    """Добавление книги в библиотеку."""
    book = {'title': title, 'author': author}
    books.append(book)
    print(f"Книга '{title}' авторства {author} добавлена в библиотеку.")


def remove_book(title: str, author: str):
    """Удаление книги из библиотеки по названию и автору."""
    for book in books:
        if book['title'] == title and book['author'] == author:
            books.remove(book)
            print(f"Книга '{title}' авторства {author} удалена из библиотеки.")
            return
    print(f"Книга '{title}' авторства {author} не найдена в библиотеке.")


def search_by_title(title: str):
    """Поиск книг по названию."""
    found_books = [book for book in books if title.lower()
                   in book['title'].lower()]
    if found_books:
        print(f"Найдено {len(found_books)} книга(и) с названием '{title}':")
        for book in found_books:
            print(f" - {book['title']} автор: {book['author']}")
    else:
        print(f"Книги с названием '{title}' не найдены.")


def search_by_author(author: str):
    """Поиск книг по автору."""
    found_books = [book for book in books if author.lower()
                   in book['author'].lower()]
    if found_books:
        print(f"Найдено {len(found_books)} книга(и) автора {author}:")
        for book in found_books:
            print(f" - {book['title']} автор: {book['author']}")
    else:
        print(f"Книги автора {author} не найдены.")


def display_books():
    """Вывод всех книг в библиотеке."""
    if books:
        print("Список всех книг в библиотеке:")
        for book in books:
            print(f" - {book['title']} автор: {book['author']}")
    else:
        print("Библиотека пуста.")


# Пример использования
# Добавление книг
add_book("Война и мир", "Лев Толстой")
add_book("Преступление и наказание", "Фёдор Достоевский")
add_book("Мастер и Маргарита", "Михаил Булгаков")

# Вывод всех книг
display_books()

# Поиск по названию
search_by_title("Война")

# Поиск по автору
search_by_author("Толстой")

# Удаление книги
remove_book("Преступление и наказание", "Фёдор Достоевский")

# Вывод всех книг после удаления
display_books()
```

#### Задача 2: Базовая статистика

Напишите программу для вывода моды, медианы и среднего

```python
# Решение
def calculate_statistics(numbers):
    # Среднее
    mean = sum(numbers) / len(numbers)

    # Мода (самое частое число)
    frequency = {}
    for num in numbers:
        frequency[num] = frequency.get(num, 0) + 1
    mode = [num for num, count in frequency.items() if count ==
            max(frequency.values())]

    # Медиана
    numbers.sort()
    n = len(numbers)
    if n % 2 == 0:
        median = (numbers[n // 2 - 1] + numbers[n // 2]) / 2
    else:
        median = numbers[n // 2]

    return mean, median, mode


# Ввод чисел пользователем
input_numbers = input("Введите числа, разделенные пробелом: ").split()
numbers = list(map(int, input_numbers))

# Получаем статистику
mean, median, mode = calculate_statistics(numbers)

# Выводим результаты
print(f"Среднее: {mean}")
print(f"Медиана: {median}")
print(f"Мода: {mode}")
```

#### Задача 3: Анализ текста

Напишите программу, которая анализирует текст, введенный пользователем, и выводит количество слов, количество уникальных слов и самое частое слово.

```python
# Подсказка: .автсежонм ьтсонщом етйатичсоП .аксипс зи овтсежонм метаз А .икортс зи косипс етйадзоС .ртсигер йинжин в етидевереП .волеборп еморк ,ыловмис еичорп и икчот ,ытяпаз есв етиребУ
s = """
Lorem Ipsum is simply dummy text of the printing and typesetting industry.
Lorem Ipsum has been the industry's standard dummy text ever since the 1500s,
when an unknown printer took a galley of type and scrambled it to make a
type specimen book. It has survived not only five centuries, but also the
leap into electronic typesetting, remaining essentially unchanged.
It was popularised in the 1960s with the release of Letraset sheets
containing Lorem Ipsum passages, and more recently with desktop publishing
software like Aldus PageMaker including versions of Lorem Ipsum.
"""

```

```python
# Решение
def analyze_text(text):
    # Приводим текст к нижнему регистру и разбиваем на слова
    words = text.lower().split()

    # Убираем знаки препинания из слов
    words = [''.join([char for char in word if char.isalnum()])
             for word in words]

    # Количество слов
    word_count = len(words)

    # Используем множество для подсчета уникальных слов
    unique_words = set(words)
    unique_words_count = len(unique_words)

    # Находим самое частое слово
    word_frequency = {}
    for word in words:
        word_frequency[word] = word_frequency.get(word, 0) + 1

    # Преобразуем частоту в множество
    max_frequency = max(word_frequency.values())

    # Ищем слова с максимальной частотой
    most_common_words = {word for word,
                         freq in word_frequency.items() if freq == max_frequency}

    return word_count, unique_words_count, most_common_words, max_frequency

# Анализ текста
word_count, unique_words_count, most_common_words, max_frequency = analyze_text(s)

# Вывод результатов
print(f"Количество слов: {word_count}")
print(f"Количество уникальных слов: {unique_words_count}")
print(f"Самое частое слово: {most_common_words} (встречается {max_frequency} раз)")
```
