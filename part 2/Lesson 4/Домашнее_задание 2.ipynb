### Задание 1: Анализ данных о студентах с использованием NumPy


**Описание:**\
У вас есть данные о студентах, включающие оценки по различным предметам. Используйте библиотеку NumPy для анализа этих данных.

```python
students_data = {
    'math': [85, 78, 92, 70, 88],
    'physics': [90, 82, 76, 85, 89],
    'informatics': [88, 92, 80, 87, 90]
}

students_data
```

**Задание:**
* Найдите среднюю оценку по каждому предмету.
* Найдите медианную оценку по каждому предмету.
* Вычислите стандартное отклонение по каждому предмету.
* Определите предмет с самой высокой средней оценкой.

```python
#решение

import numpy as np


def analyze_students_data(students_data):
    results = {}

    for subject, scores in students_data.items():
        mean_score = np.mean(scores)
        median_score = np.median(scores)
        std_dev = np.std(scores)

        results[subject] = {
            'Средняя оценка': mean_score,
            'Медианная оценка': median_score,
            'Стандартное отклонение': std_dev
        }

    # Определяем предмет с самой высокой средней оценкой
    highest_avg_subject = max(
        students_data, key=lambda sub: np.mean(students_data[sub]))

    # Вывод результатов
    for subject, stats in results.items():
        print(f"{subject.capitalize()}:")
        for stat, value in stats.items():
            print(f"  {stat}: {value:.2f}")
        print()

    print(
        f"Предмет с самой высокой средней оценкой: {highest_avg_subject.capitalize()}")


analyze_students_data(students_data)

```

### Задание 2: Анализ данных о продажах с использованием Pandas

**Описание:**\
Используйте библиотеку Pandas для анализа данных о продажах товаров. У вас есть данные о продажах товаров за последний год.

```python
sales_data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
    'Sales': [15000, 17000, 16000, 18000, 14000, 19000, 22000, 21000, 23000, 25000, 24000, 26000]
}
sales_data
```

**Задание:**
* Найдите общий объем продаж за год.
* Найдите средний объем продаж в месяц.
* Найдите месяц с наибольшим и наименьшим объемом продаж.
* Постройте линейный график изменения объема продаж по месяцам (Matplotlib).

```python
# Решение

import pandas as pd
import matplotlib.pyplot as plt


def analyze_sales(sales_data):
    df = pd.DataFrame(sales_data)

    # Общий объем продаж за год
    total_sales = df['Sales'].sum()

    # Средний объем продаж в месяц
    average_sales = df['Sales'].mean()

    # Месяцы с наибольшими и наименьшими продажами
    max_sales_month = df.loc[df['Sales'].idxmax(), 'Month']
    min_sales_month = df.loc[df['Sales'].idxmin(), 'Month']
    
    # Вывод результатов
    print(f"Общий объем продаж за год: {total_sales}")
    print(f"Средний объем продаж в месяц: {average_sales:.2f}")
    print(f"Месяц с наибольшими продажами: {max_sales_month}")
    print(f"Месяц с наименьшими продажами: {min_sales_month}")

    # График продаж по месяцам
    plt.figure(figsize=(10, 5))
    plt.plot(df['Month'], df['Sales'], marker='o', linestyle='-', color='g')
    plt.xticks(rotation=45)
    plt.xlabel('Месяц')
    plt.ylabel('Продажи')
    plt.title('Динамика продаж по месяцам')
    plt.grid()
    plt.show()


# Вызов функции
analyze_sales(sales_data)
```

### Задание 3: Задание на визуализацию данных с использованием Seaborn

**Описание:**\
Используйте готовый набор данных из библиотеки Seaborn для создания различных типов графиков. \
Ваша задача — загрузить датасет ***tips*** из готового набора Seaborn и построить несколько базовых графиков для анализа данных.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips') # используйте строго данный датасет

df.head()
```

Постройте следующие графики:
* Линейный график
* Столбчатый график
* Гистограмма
* Диаграмма рассеяния

Условия:
* Параметры визуализации на ваш выбор
* Графики должны нести смысловую нагрузку

```python
# Решение

# Настройка размеров фигуры
plt.figure(figsize=(15, 10))

# Линейный график (средний счет по дням)
plt.subplot(2, 2, 1)
sns.lineplot(x='day', y='total_bill', data=df,
             estimator='mean', errorbar=None, marker='o', color='b')
plt.title('Средний счет по дням недели')

# Столбчатый график (сумма чаевых по дням)
plt.subplot(2, 2, 2)
sns.barplot(x='day', y='tip', data=df, estimator='sum',
            hue='day', legend=False, palette='coolwarm')
plt.title('Сумма чаевых по дням недели')

# Гистограмма (распределение суммы счета)
plt.subplot(2, 2, 3)
sns.histplot(df['total_bill'], bins=20, kde=True, color='g')
plt.title('Распределение суммы счета')

# Диаграмма рассеяния (зависимость чаевых от суммы счета)
plt.subplot(2, 2, 4)
sns.scatterplot(x='total_bill', y='tip', data=df, hue='sex',
                style='smoker', palette='viridis')
plt.title('Зависимость чаевых от суммы счета')

```
